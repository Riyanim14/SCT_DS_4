```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set random seed for reproducibility
random.seed(42)
```

```python

# Create a synthetic dataset
num_records = 1000
road_conditions = ['Dry', 'Wet', 'Icy', 'Snowy']
weather_conditions = ['Clear', 'Rain', 'Fog', 'Snow']
times_of_day = ['Morning', 'Afternoon', 'Evening', 'Night']

# Generate random data
accident_data = {
    'Accident_ID': range(1, num_records + 1),
    'Road_Condition': [random.choice(road_conditions) for _ in range(num_records)],
    'Weather_Condition': [random.choice(weather_conditions) for _ in range(num_records)],
    'Time_of_Day': [random.choice(times_of_day) for _ in range(num_records)],
    'Accident_Severity': [random.randint(1, 5) for _ in range(num_records)],  # 1 to 5 scale
    'Location': [f'Location_{random.randint(1, 20)}' for _ in range(num_records)],
}
```

```python

# Create DataFrame
accident_df = pd.DataFrame(accident_data)

# Save to CSV
accident_df.to_csv('traffic_accident_data.csv', index=False)
```

```python
# Display the head of the dataset
print(accident_df.head())

```

```python
# Enhanced CountPlot: Accident Hotspots by Location and Severity
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style='whitegrid')

# Sort locations by total accidents for better readability
location_order = (
    accident_df['Location']
    .value_counts()
    .sort_values(ascending=False)
    .index
)

# Set up the figure
plt.figure(figsize=(16, 8))
sns.countplot(
    data=accident_df,
    x='Location',
    hue='Accident_Severity',
    order=location_order,
    palette='coolwarm'
)

# Titles and labels with improved formatting
plt.title(' Accident Hotspots by Location and Severity', fontsize=18, weight='bold')
plt.xlabel(' Location', fontsize=14)
plt.ylabel(' Number of Accidents', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=11)

# Legend improvements
plt.legend(
    title=' Severity Level',
    title_fontsize=12,
    fontsize=10,
    loc='upper right',
    bbox_to_anchor=(1.12, 1)
)

# Add gridlines for clarity
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Layout and save
plt.tight_layout()
plt.savefig('accident_hotspots_enhanced.png', dpi=300, bbox_inches='tight')
plt.show()
```
