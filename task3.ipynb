```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\riyan\OneDrive\Desktop\TASK 3\bank.csv")  # no sep argument
```

```python
# Clean column names
df.columns = df.columns.str.strip()

# Print column names to confirm
print("Columns:", df.columns.tolist())
```

```python
# Encode categorical variables
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Now
X = df.drop('deposit', axis=1)
y = df['deposit']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train classifier
clf = DecisionTreeClassifier(max_depth=5, random_state=42)
clf.fit(X_train, y_train)
```

```python
# Predictions
y_pred = clf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(24, 12))  # Bigger canvas
plot_tree(
    clf,
    feature_names=X.columns,
    class_names=label_encoders['deposit'].classes_,
    filled=True,
    rounded=True,              # Rounded boxes
    proportion=True,           # Show proportion instead of counts
    precision=2,               # Control decimal places
    fontsize=10,               # Adjust font size
    impurity=True              # Show Gini/Entropy
)
plt.title("ðŸŒ³ Enhanced Decision Tree - Bank Term Deposit Prediction", fontsize=16)
plt.tight_layout()
plt.show()
```
